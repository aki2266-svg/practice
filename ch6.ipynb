```python
f=open('compare1.txt')
f1=open('compare2.txt')
s1=f.read()
s2=f1.read()
line,pos=1,1
for i in range(len(s1)):
    if s1[i]==s2[i]:
        pos+=1
        if s1[i]=="\n":
            line+=1
            pos=1
    else:
        print(f"line:{line},pos:{pos}")
        pos+=1
f.close()
f1.close()
```
